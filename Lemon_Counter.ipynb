```
# Importing libraries
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import time
from google.colab.patches import cv2_imshow

# defining lemonCounter function
def lemonCounter(path):
  # starting time of execution
  start_time = time.time()

  image = cv2.imread(path, 1)

  # Color detection for different green modes for lemons
  lower, upper = ([25,52,72], [102,255,255])  # pixels other range get dark 
  lower = np.array(lower, dtype = "uint8")
  upper = np.array(upper, dtype = "uint8")
  kernal = np.ones((7,7), 'uint8')
  mask = cv2.inRange(image, lower, upper)
  green_mask = cv2.dilate(mask, kernal)
  output = cv2.bitwise_and(image, image, mask = green_mask)
  #gray correction of image
  gray = cv2.cvtColor(output, cv2.COLOR_BGR2GRAY)
  gray = np.array(255 * (gray / 255) ** 1.3, dtype='uint8')

  # filtering image, applying gaussian blur with kernel size (7,7)
  gray = cv2.GaussianBlur(gray, (7,7), cv2.BORDER_ISOLATED)
  # segmentation using otsu thresholding for seperation of background and object
  ret3,gray = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU) 
  gray = cv2.GaussianBlur(gray, (7,7), cv2.BORDER_ISOLATED) #filtering again for removing unnecessary area

  # morphological operations for processing the image for elliptical objects
  kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
  gray = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel, iterations=3)

  # using findContours function for finding white objects from black background
  count, hier = cv2.findContours(gray, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
  num = str(len(count))
  end_time = time.time()
  time_taken = end_time-start_time

  return image, time_taken, num
  
```

```
paths = ['lemons1.jpeg', 'lemon1.jpg', 'lemon3.jpg', 'lemon4.jpg', 'lemon5.jpg', 'lemon7.jpg']
for path in paths:
  img, time_taken, num = lemonCounter(path)
  plt.figure(figsize = (8,8))
  plt.axis('off')
  plt.imshow(img)
  plt.show()
  print("Time taken for execution:", time_taken)
  print("Total number of lemons detected:", num)
```
